For the rotating mass we can write $x,y$ and $z$ as

$x=r\cos(\theta) \hspace{0.5cm}\Rightarrow \dot x=\dot r\cos(\theta)-r\sin(\theta)\dot \theta$


$y=r\sin(\theta)\hspace{0.5cm}\Rightarrow \dot y=\dot r\sin(\theta)+r\cos(\theta) \dot \theta$


$z=0\hspace{1.6cm}\Rightarrow \dot z=0$

For the hanging mass we can write $x,y$ and $z$ as

$x=0\hspace{2.4cm}\Rightarrow \dot x=0$

$y=0\hspace{2.4cm}\Rightarrow \dot y=0  $

$z=-(l_0-r)\hspace{0.5cm}\Rightarrow \dot z=\dot r$

Kinetic energy of the system

$T=\frac{1}{2}m\big[\dot x^2+\dot y^2+\dot z^2\big]=\frac{1}{2}m\big[\dot r^2+r^2\dot \theta^2+\dot r^2\big]=\frac{1}{2}m\big[2\dot r^2+r^2\dot \theta^2\big]$

Potential energy of the system

$V=-mg(l_0-r)$

Lagragian of the system

$L=T+V=\frac{1}{2}m\big[2\dot r^2+r^2\dot \theta^2\big]+mg(l_0-r)$

Now 

$p_r=\frac{\partial L}{\partial \dot r}=2m\dot r$

$p_{\theta}=\frac{\partial L}{\partial \dot \theta}=mr^2\dot \theta$

Now the Hamiltonian can be written as

$H=T+V=\frac{1}{2}m\big[2\dot r^2+r^2\dot \theta^2\big]-mg(l_0-r)$

$H=\frac{p_r^2}{4m}+\frac{p_{\theta}^2}{2mr^2}-mg(l_0-r)$

Now the equations of motion are

$\dot r=\frac{\partial H}{\partial p_r}=\frac{p_r}{2m}$

$\dot \theta=\frac{\partial H}{\partial p_{\theta}}=\frac{p_{\theta}}{mr^2}$

and

$-\dot p_r=\frac{\partial H}{\partial r}=-\frac{p_{\theta}^2}{mr^3}+mg$

$-2m\ddot r=-\frac{m^2r^4\dot \theta^2}{mr^3}+mg$

$-2m\ddot r=-mr\dot \theta^2+mg$

$\ddot r=\frac{r\dot \theta^2-g}{2}$

Also

$-\dot p_{\theta}=\frac{\partial H}{\partial \theta}=0$

This suggest that angular mementum in $\theta$ direction is a conserved quantity

Also we can wrote

$-\frac{d}{dt}(p_{\theta})=-\frac{d}{dt}(mr^2\dot \theta)=0 \hspace{0.5cm}\Rightarrow r^2\ddot \theta+2r\dot r\dot \theta= 0\hspace{0.5cm}\Rightarrow r \ddot \theta+2\dot r \dot \theta=0$


Total energy of the system is conserved

$\frac{\partial H}{\partial t}=2m\dot r\ddot r+mr\dot r \dot \theta^2+mr^2\dot \theta \ddot \theta+mg\dot r=m\dot r(r\dot \theta^2-g)+mr\dot r \dot \theta^2+mr^2\dot \theta (-\frac{2\dot r \dot \theta}{r})+mg\dot r=0$

At the equilibrium position the tension in the string is equal to the centrifugal force of the rotationg mass 

Now

$r\dot \theta^2=g$ is the equilibrium condition


Now

$\ddot r=\frac{r\dot \theta^2-g}{2}$

$\ddot \theta=-\frac{2\dot r \dot \theta}{r}$


Let's use the substitute

$R=\dot r$

$\chi=\dot \theta$

So

$\dot R=\frac{r\chi^2-g}{2}$

$\dot \chi=-\frac{2R \chi}{r}$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
r=1

def f_int(y,t):  #define the equations of motion
    
    R,chi=y
   
    return [(r*chi**2-9.8)/2,-2*R*chi/r]
```

```python
r_0 = np.array([0,1])   # initial conditions r_dot=0 and theta_dot =1
t = np.linspace(0.0, 10.0, 1000)   
```

```python
r_sol = odeint(f_int, r_0, t)  #solve eqations of motion
```

```python
plt.plot(t, r_sol[:, 0])
plt.xlabel('t')
plt.ylabel('$r$')
plt.show()#plot r

plt.plot(t, r_sol[:, 1])
plt.xlabel('t')
plt.ylabel('$\Theta$')
plt.show()#plot theta

```

These are the oscillations of $r$ and $\theta$ in the equilibrium position
