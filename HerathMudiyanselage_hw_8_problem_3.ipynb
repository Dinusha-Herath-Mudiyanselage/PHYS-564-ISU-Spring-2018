For the spring pendulum

Kinetic energy term can be written as

$T=\frac{1}{2}m\big[{\dot r}^2+r^2 \dot \theta \hspace{1mm} ^2]$

and potential energy

$V=\frac{1}{2}k(r-a)^2-mgr\cos(\theta)$ where $a$ is the natural length of the spring.

So the Lagragian is 

$L=T-V=\frac{1}{2}m\big[{\dot r}^2+r^2 \dot \theta \hspace{1mm} ^2]-\frac{1}{2}k(r-a)^2+mgr\cos(\theta)$

So the equations of motion are

$\frac{d}{dt}\big(\frac{\partial L}{\partial \dot r}\big)-\frac{\partial L}{\partial  r}=0$

$m\ddot r-mr\dot \theta \hspace{1mm}^2+k(r-a)-mg\cos(\theta)=0$

$\ddot r=r\dot \theta \hspace{1mm}^2+g\cos(\theta)-\frac{k}{m}(r-a)$

$\frac{d}{dt}\big(\frac{\partial L}{\partial \dot \theta}\big)-\frac{\partial L}{\partial  \theta}=0$

$mr^2\ddot \theta +2mr\dot r\dot \theta+mgr\sin(\theta)=0$

$r\ddot \theta +2\dot r\dot \theta+g\sin(\theta)=0$

For the equilibrium configuration potential enrgy should be a minimum

$\frac{\partial V}{\partial r}=0$ and $\frac{\partial V}{\partial \theta}=0$

Since $V=\frac{1}{2}k(r-a)^2-mgr\cos(\theta)$

$k(r-a)-mg\cos(\theta)=0$ and 

$mgr\sin(\theta)=0$

With the small angle approximation $\cos(\theta)$~$1$ and $\sin(\theta)$~$\theta$ the equations of motion becomes 

$\ddot r=r\dot \theta \hspace{1mm}^2+g-\frac{k}{m}(r-a)$

$r\ddot \theta +2\dot r\dot \theta+g\theta=0$

Now using the equilibrium configurations

$k(r-a)=mg$ 

So

$\ddot r=r\dot \theta \hspace{1mm}^2$

$r\ddot \theta +2\dot r\dot \theta+g\theta=0$


........................................

and next $g\theta=0$

$\ddot r=r\dot \theta \hspace{1mm}^2+g-\frac{k}{m}(r-a)$

$r\ddot \theta +2\dot r\dot \theta=0$


Now let's solve the equations

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

Now consider the first set of equations

$\ddot r=r\dot \theta \hspace{1mm}^2$

$r\ddot \theta +2\dot r\dot \theta+g\theta=0$

now $r=\frac{mg}{k}+a$ and $\dot r=0$

$\ddot \theta=-\frac{g}{r}\theta$ vibration frequency is $f=\frac{1}{2\pi}\sqrt{\frac{g}{r}}=\frac{1}{2\pi}\sqrt{\frac{9.8}{14.8}}=0.13s^{-1}$ period$=7.72s$

The system acts like a normal pendulum , no vibrations along r direction

For $m=1kg, k= 1 Nm^{-1}$ and $a=5m$

$\ddot\theta + \frac{g}{r}\theta=0$

$\frac{d\theta_0}{dt}=-\frac{g}{r}\theta_1$

$ \frac{d\theta_1}{dt}=\theta_0$  

```python
m = 1
g = 9.8
k = 1
a = 5

r=(m*g/k)+a

def func(theta,t):
    return [-(g/r) *theta[1],theta[0]]
```

```python
y_0 = np.array([np.pi/10,0])
t = np.linspace(0.0, 20.0, 100)
```

```python
y_sol = odeint(func, y_0, t)
```

```python
plt.plot(t, y_sol[:, 0]) #plot theta
plt.show()
print('Period is',t[np.argmin(y_sol[:, 0])]*2)# find the period from graph
```

Let's plot $\ddot r$

```python
plt.plot(t, r*y_sol[:, 1]**2) #plot r_double dot
plt.show()
```

Next solve the 2 nd set of equations

$\ddot r=r\dot \theta \hspace{1mm}^2+g-\frac{k}{m}(r-a)$

$r\ddot \theta +2\dot r\dot \theta=0$

This case is correspond to $\theta=0$ case
 so there is no $\dot \theta$ becasue there is no swing in the pendiulum but oscillating up and down.
 
 Hence
 
 
 $\ddot r=-\frac{k}{m}r+g+\frac{ka}{m}$
 
  vibration frequency is $f=\frac{1}{2\pi}\sqrt{\frac{k}{m}}=\frac{1}{2\pi}\sqrt{\frac{1}{1}}=0.16s^{-1}$ period$=6.28s$


$\frac{d^2r}{dt}+\omega^2r=g+\omega^2 a$

$\frac{dr_0}{dt}=-r_1$

$ \frac{dr_1}{dt}=r_0-14.8$  

```python
def func(r,t):       
    return [-k/m*r[1],r[0]-14.8]
```

```python
y_0 = np.array([15.0,0.0])        
t = np.linspace(0.0, 20.0, 100)
```

```python
y_sol = odeint(func, y_0, t)   
```

```python
plt.plot(t, y_sol[:, 0])
plt.show()
print('Period is',t[np.argmin(y_sol[:50, 0])]-t[np.argmin(y_sol[:40, 0])])# find the period from graph
```
