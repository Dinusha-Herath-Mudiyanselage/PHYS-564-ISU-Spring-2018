Poincare map for the pendulum

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def egy(p, E, F, G):
    return np.arccos((0.5*G*p**2-E)/F)
```

```python
def p_max(E, F, G):
    return np.sqrt(2.0*(E+F)/G)
def p_min(E, F, G):
    return np.sqrt(2.0*(E-F)/G)
```

```python
F=1
G=1
```

```python
num_pnts = 300
eps = 1e-6

e_vals = np.array([-0.99, -0.5, 0.0, 0.3, 0.99])

for e_v in e_vals:
    p_m = p_max(e_v, F, G)
    p_var = np.linspace(-p_m+eps, p_m-eps, num_pnts)
    plt.plot(egy(p_var, e_v, F, G), p_var)
    plt.plot(-egy(p_var, e_v, F, G), p_var)

e_vals = np.array([1.01, 1.3, 1.5])

for e_v in e_vals:
    p_mx = p_max(e_v, F, G)
    p_mn = p_min(e_v, F, G)
    p_var = np.linspace(p_mn+eps, p_mx-eps, num_pnts)
    plt.plot(egy(p_var, e_v, F, G), p_var)
    plt.plot(-egy(p_var, e_v, F, G), p_var)
    plt.plot(egy(p_var, e_v, F, G), -p_var)
    plt.plot(-egy(p_var, e_v, F, G), -p_var)

plt.title('Poincare Map for the simple pendulum')
plt.xlabel(r'q-coordinate')
plt.ylabel(r'p-coordinate')
plt.show()
```
