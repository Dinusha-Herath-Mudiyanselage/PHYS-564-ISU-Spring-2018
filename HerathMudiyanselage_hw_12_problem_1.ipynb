Map is defined by $F(x)= a\sin(\pi x)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

Define the map

```python
def sin_map(x, a=0.5):
    
    return a*np.sin(np.pi*x)
```

Let's plot for $a=1/24$

Now $x_n=\frac{1}{24}\sin(\pi x_n)$

Fixed point is $x_n=0$

Fixed point is obtained by geting the intesection point of two graphs $x_n$ and 

$\frac{1}{24}\sin(\pi x_n)$

```python
a_param = 1/24
ini_val = 0.1

n_iter = 50
x_iter = np.zeros(n_iter)

z = ini_val
for ind in range(n_iter):
    z = sin_map(z, a_param)
    x_iter[ind] = z

plt.plot(range(n_iter), 0*np.ones(n_iter),'--')
plt.plot(range(n_iter), x_iter)
plt.show()
```

There is no bifurcation in the plot.
