Lagragian for the pendulum

$L=T-V$

$T=\frac{1}{2}ml^2(\dot\theta)^2  \hspace{2cm} V=-mglcos(\theta)$

$  L = \frac{1}{2}ml^2(\dot\theta)^2+mglcos(\theta)   \hspace{1cm}$ where $l$ is the length of the pendulum

Equation of motion can be obtained from

$\frac {d}{dt}(\frac{\partial L}{\partial \dot\theta})-(\frac{\partial L}{\partial \theta})=0 $  

$\frac{\partial L}{\partial \dot\theta}=ml^2\dot\theta   \hspace{1cm}$ and $\hspace{1cm}  \frac{\partial L}{\partial \theta}=-mglsin(\theta)$


Hence the equation of motion is

$ml^2\ddot\theta+mglsin(\theta)=0$

$\ddot\theta+(\frac{g}{l})sin(\theta)=0$

Boundary conditions

$\frac{d^2\theta}{dt^2}+(\frac{g}{l})sin(\theta)=0 $ with $ \theta(0)=\frac{\pi}{3}$ and $l\dot\theta(0)=0$ and $g=9.8$ m/s/s

We can write the above differential eqaution with two other eqautions.

$\frac{d\theta_0}{dt}=-\theta_1$

$\frac{d\theta_1}{dt}=\frac{g}{l}sin(\theta_0) $ with $ \theta_0(0)=\frac{\pi}{3}$ and $\theta_1(0)=0$

First import all the packages necessary for the program.

```python
import numpy as np                     #import the numpy package
from scipy.integrate import odeint     #import the odient function from scipy.integrate module
import matplotlib.pyplot as plt        #import the matplotlib package
```

Define the differential equation

```python
l=9.8 #length in m

def f_int(theta,t):             #define ODE
    return [-theta[1],(9.8/l)*np.sin(theta[0])]
```

Then implement initial conditions

```python
y_0 = np.array([np.pi/3,0])      # an array with initial conditions pi/3 and 0
t = np.linspace(0.0, 20.0, 100)  # create a line space array with 100 eqaul space points between 0 and 20
```

After that solve the differential equation

```python
y_sol = odeint(f_int, y_0, t)    #solve the diffential equation with boundary conditions
```

Finally plot the solution

```python
plt.plot(t, y_sol[:, 0])
plt.title('Solution of the differential equation')
plt.show()
```

With the small angle approximation the ODE reads as

$\frac{d^2\theta}{dt}+\omega^2\theta=0$ where $\omega=\sqrt\frac{g}{l}$ since $\omega=1$

$\frac{d^2\theta}{dt}+\theta=0$

The analytical solution is $\theta(t)=Asin(t)+Bcos(t)$ with boundray conditions $ \theta(0)=\frac{\pi}{3}$ and $l\dot\theta(0)=0$

Solving for $A$ and $B$ gives $A=0$ and $B=\frac{\pi}{3}$

Hence  $\theta(t)=\frac{\pi}{3}cos(t)$



Lets plot the analytical solution

```python
def f_int_a(t):             #define solution
    return np.pi/3*np.cos(t)
```

```python
y_analytical=f_int_a(t)
```

```python
plt.plot(t, y_sol[:, 0])
plt.plot(t,y_analytical)
plt.title('Different solutions of the differential equation')
plt.show()
```

Confirms that small angle approximation is wrong.

Next Calculate the error between these two solutions

```python
plt.plot(t, 100*np.abs((y_sol[:, 0]-y_analytical)/y_analytical))
plt.show()
```
