the lagragian L=1/2(mx'^2-kx^2) ............. x' is the time derivative of x coordinate 

the equation of motion is mx''+kx=0

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f_int(y,t):
    return [-y[1],y[0]]
```

```python
y_0 = np.array([0.0, -2.0])
t = np.linspace(0.0, 100.0, 100)
```

```python
y_sol = odeint(f_int, y_0, t)
```

```python
plt.plot(t, y_sol[:, 0])
plt.show()
```
