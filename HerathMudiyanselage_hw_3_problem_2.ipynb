Lagragian for the system

$L=T-V$

$T=\frac{1}{2}m(\dot y)^2  \hspace{2cm} V=\frac{1}{2}k(y-l_0)^2-mgy$

$  L = \frac{1}{2}m(\dot y)^2-\frac{1}{2}k(y-l_0)^2+mgy   \hspace{1cm}$ where $l_0$ is the unstreched length of the spring

Equation of motion can be obtained from

$\frac {d}{dt}(\frac{\partial L}{\partial \dot y})-(\frac{\partial L}{\partial y})=0 $  

$\frac{\partial L}{\partial \dot y}=m\dot y   \hspace{1cm}$ and $\hspace{1cm}  \frac{\partial L}{\partial \theta}=-k(y-l_0)+mg$

Hence the equation of motion is

$m\ddot y+k(y-l_0)-mg=0$

$\ddot y+(\frac{k}{m})y=g+\frac{k}{m}l_0$

Lets solve the second order non homogenous differential eqaution

First introduce $\omega^2=\frac{k}{m}$

Now $\ddot y+\omega^2y=g+\omega^2 l_0$

Complementary solution for the homogenous eqaution $\ddot y+(\frac{k}{m})y=0$ is

$y_c(t)=Asin(\omega t)+Bcos(\omega t)$

 Patricular solution for the non homogenous eqaution $\ddot y+(\frac{k}{m})y=g+\omega^2 l_0$ is

Let $\dot y=Dy$ and $ \ddot y=D^2y$

$(D^2+\omega^2)y=g+\omega^2 l_0$

$y=\frac{g}{D^2+\omega^2}+\frac{l_0}{D^2+\omega^2}$

$y=\frac{g}{D^2+\omega^2}\exp(0t) +\frac{l_0\omega^2}{D^2+\omega^2}\exp(0t)$

Now putting $D=0$

We have the paticular solution

$y_p(t)=\frac{g}{\omega^2}+l_0=\frac{mg}{k}+l_0$

Now the genaral solution is

$y(t)=y_c(t)+y_p(t)=\frac{mg}{k}+l_0+Asin(\omega t)+Bcos(\omega t)$

Lets apply the boundary conditions $y (t=0)=y_0$ and $\dot y(t=0)=v_0$

$y_0=\frac{mg}{k}+l_0+B$

$B=y_0-(\frac{mg}{k}+l_0)$

Now lets define $\Delta y_0=y_0-(\frac{mg}{k}+l_0)$

Hence $B=\Delta y_0$

And $v_0=A\omega$ hence $A=\frac{v_0}{\omega}$

Now the general solution becomes

$y(t)=\frac{mg}{k}+l_0+\Delta y_0cos(\omega t)+\frac{v_0}{\omega}sin(\omega t)$

The value $(\frac{mg}{k}+l_0)$ is the total displacement from the suspended point when the system is not in oscillation. $l_0$ is the natural length of the spring and $\frac{mg}{k}$ is the distance the spring stretched once the mass is attached to the spring. Once the system oscillates  $\Delta y_0 =y_0-(\frac{mg}{k}+l_0)$ is the difference between the  equiblirium point and the inital displacement of the system (t=0).

$v_0$ is the initail velocity of the mass (veleocity at t=0)

$\omega=\sqrt\frac{k}{m}$ here $\omega$ is the angular frequecy of the system which depends on spring constant and the hanged mass



Lets solve the equation of motion analytically

$\frac{d^2y}{dt}+\omega^2y=g+\omega^2 l_0$

Lets assume $l_0$=1 m ,$k$= 1 N/m , $\Delta y_0$= 0.1 m and $v_0$=0 and $m$=1 kg also $g=9.8$ m/s/s

So the equation reduces to

$\frac{d^2y}{dt}+y=10.8$ with boundary conditions $y(0)=\Delta y_0+(\frac{mg}{k}+l_0) =10.9$ m and $y'(0)=v_0 = 0$ m/s 

We can write the above differential eqaution with two other eqautions.

$\frac{dy_0}{dx}=-y_1$

$ \frac{dy_1}{dx}=y_0-10.8$  

with $y_0(0)=10.9$ and $y_1(0)=0$

First import all the packages necessary for the program.

```python
import numpy as np                     #import the numpy package
from scipy.integrate import odeint     #import the odient function from scipy.integrate module
import matplotlib.pyplot as plt        #import the matplotlib package
```

Define the differential equation

```python
def f_int(y,t):       #define ODE
    return [-y[1],y[0]-10.8]
```

Then implement initial conditions

```python
y_0 = np.array([10.9,0.0])        # an array with initial conditions 10.9 and 0
t = np.linspace(0.0, 20.0, 100)   # create a line space array with 100 eqaul space points between 0 and 20
```

After that solve the differential equation

```python
y_sol = odeint(f_int, y_0, t)    #solve the diffential equation with boundary conditions
```

Finally plot the solution

```python
plt.plot(t, y_sol[:, 0])
plt.title('Solution of the differential equation')
plt.show()
```

Lets Compare the annalytical solution and the numerical solution 

Analytical solution is $y(t)=10.8+0.1cos(t)$

```python
def f_int_a(t):       #define ODE
    return 10.8+0.1*np.cos(t)
```

```python
y_analytical=f_int_a(t)
```

```python
plt.plot(t, y_sol[:, 0])
plt.plot(t,y_analytical)
plt.title('Different solutions of the differential equation')
plt.show()
```

Both solutions are identical

Next calculate the error between exact and numerical solutions

```python
plt.plot(t, 100*np.abs((y_sol[:, 0]-y_analytical)/y_analytical))
plt.show()
```
