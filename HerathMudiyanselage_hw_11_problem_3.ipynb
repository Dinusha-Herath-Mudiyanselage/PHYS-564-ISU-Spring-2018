First import all the necessary pacakges

```python
import numpy as np
import matplotlib.pylab as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

Now the equations for the Rossler attractor is

$\frac{dx}{dt}=-y-z$

$\frac{dy}{dt}=x+ay$

$\frac{dz}{dt}=b+z(x-c)$

```python
def fun(y, t, a, b, c):
    """
    Integration function
    """
    f1 = -y[1]-y[2]
    f2 = y[0]+a*y[1]
    f3 = b+y[2]*(y[0]-c)
    
    return np.array([f1, f2, f3])
```

```python
def solv(y0, t, a, b, c):
    return odeint(fun, y0, t, args=(a, b, c))
```

Lets solve for $c=4.0$

```python
a = 0.1
b = 0.1
c = 4.0
t_end = 8.0*np.pi
num_points = 100
y0 = np.array([1.0,1.0, 0.0])
temp = np.linspace(0.0, t_end, num_points)
sol = solv(y0, temp, a, b, c)
plt.plot(temp, sol)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


X = sol[:, 0]
Y = sol[:, 1]
Z = sol[:, 2]

ax.plot(X,Y,Z)
plt.show()
```

2D plot in x-y plane

```python
plt.plot(sol[:, 0], sol[:, 1])
plt.show()
```

Now lets look at the graphs by increasing the time

```python
temp = np.linspace(0.0, t_end*3, 3*num_points)
sol = solv(y0, temp, a, b, c)
plt.plot(temp, sol)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


X = sol[:, 0]
Y = sol[:, 1]
Z = sol[:, 2]

ax.plot(X,Y,Z)
plt.show()
```

```python
plt.plot(sol[:, 0], sol[:, 1])
plt.show()
```

The orbit is a closed cycle for $c=4.0$ we have a limit cycle
