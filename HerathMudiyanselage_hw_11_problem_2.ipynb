x(t) = A\sin(a t + \delta)
$

$
y(t) = B\sin(b t)
$

```python
import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
```

Lets define the function to calculate values

```python
def liss(t, a, b, amp_a, amp_b, delt):
    """returns the Lissajous figure for the given paramters
    """
    return np.array([amp_a*np.sin(a*t+ delt), amp_b*np.sin(b*t)])
```

Now let's draw the figure for $\frac{a}{b}=\frac{\sqrt2}{\sqrt3}$ and $\delta=0$

```python
num_pnts = 2000
delt_num = 0

a= 1.0
amp_a = 1.0
b = (np.sqrt(3)/np.sqrt(2))*a
amp_b = amp_a
temp = np.linspace(0.0, 2*np.pi*20, num_pnts)
```

Let's plot the results

```python
x, y = liss(temp, a, b, amp_a, amp_b, delt_num)
plt.plot(x, y)
plt.text(-1.5, 0.7, r'$\delta =0$')
plt.axis('equal')
plt.xlim([-1, 1])
plt.show()
```

Let's plot the probability distribution

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

histogram,x_edge,y_edge=np.histogram2d(x , y, bins= 100 , range=[[-1,1],[-1,1]])
x_position, y_position = np.meshgrid(x_edge[:-1]+0.05 , y_edge[:-1] +0.05)


x_position = x_position.flatten('F')
y_position = y_position.flatten('F')
z_position = np.zeros_like(x_position)


dx = 0.1 * np.ones_like(z_position)
dy = dx.copy()
dz = hist.flatten()

ax.bar3d(x_position, y_position, z_position, dx, dy, dz, color='b', zsort='average')

plt.show()
```
