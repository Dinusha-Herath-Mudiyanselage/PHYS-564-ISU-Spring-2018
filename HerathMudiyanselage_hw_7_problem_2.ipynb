The original equations for a particle in a sphers is

$  \frac{d \theta(t)}{dt} = \pm \sqrt{ \dot{\theta}_0^2 + \dot{\varphi}_0^2 \sin^4(\theta_0)\left( \frac{1}{\sin^2(\theta_0)}-\frac{1}{\sin^2(\theta(t))} \right) } $

 $ \frac{d \varphi(t)}{dt} = \dot{\varphi}_0\frac{\sin^2(\theta_0)}{\sin^2(\theta(t))}$

 with boundary conditions $\varphi(0)=\varphi_0$ and $\theta(0)=\theta_0$. the constants $\dot{\varphi}_0$ and $\dot{\theta_0}$ are the initial angular velocities.
 
 Later we derive another set of equations using thses equations
 
 $\dot\chi(t)=\dot{\varphi}_0^2 \sin^4(\theta_0) \frac{\cos(\theta(t)}{\sin^3(\theta(t))}$

$\dot\theta(t)=\chi(t)$

$ \dot \varphi(t) = \dot{\varphi}_0\frac{\sin^2(\theta_0)}{\sin^2(\theta(t))}$

 with boundary conditions $\varphi(0)=\varphi_0$ and $\theta(0)=\theta_0$ and $\chi(0)=\dot\theta_0$ the constants $\dot{\varphi}_0$ and $\dot{\theta_0}$ are the initial angular velocities.

Now let's use the substitution $u(t)=\cos(\theta(t))$

$\dot u(t)= - \sin(\theta(t))\frac{d\theta(t)}{dt}$

$\ddot u(t)=-\cos(\theta(t))\frac{d\theta(t)}{dt}\frac{d\theta(t)}{dt} - \sin(\theta(t))\frac{d^2\theta(t)}{dt^2}$

$\ddot u(t)=-\cos(\theta(t))(\dot\theta(t))^2 - \sin(\theta(t))\dot\chi(t)$

now substituting $\dot\theta(t)$ and $\chi(t)$ into the above equation

$\ddot u(t)=-\cos(\theta(t)) \bigg[\dot{\theta}_0^2 + \dot{\varphi}_0^2 \sin^4(\theta_0)\left( \frac{1}{\sin^2(\theta_0)}-\frac{1}{\sin^2(\theta(t))} \right) \bigg] - \sin(\theta(t))\bigg[\dot{\varphi}_0^2 \sin^4(\theta_0) \frac{\cos(\theta(t)}{\sin^3(\theta(t))}\bigg]$

$\ddot u(t)=-\cos(\theta(t)) \big[\dot{\theta}_0^2 + \dot{\varphi}_0^2 \sin^2(\theta_0)\big]+ \dot{\varphi}_0^2 \sin^4(\theta_0) \frac{\cos(\theta(t)}{\sin^2(\theta(t))} - \dot{\varphi}_0^2 \sin^4(\theta_0) \frac{\cos(\theta(t)}{\sin^2(\theta(t))}$

$\ddot u(t)=-\cos(\theta(t)) \big[\dot{\theta}_0^2 + \dot{\varphi}_0^2 \sin^2(\theta_0)\big]$

$\ddot u(t)=-\big[\dot{\theta}_0^2 + \dot{\varphi}_0^2 \sin^2(\theta_0)\big]u(t)$

$\ddot u(t)=-{\Omega_0}^2 u(t)$   where ${\Omega_0}^2=\dot{\theta}_0^2 + \dot{\varphi}_0^2 \sin^2(\theta_0)$

and we can also write 

$ \dot \varphi(t) = \dot{\varphi}_0\frac{\sin^2(\theta_0)}{\sin^2(\theta(t))}=\dot{\varphi}_0\frac{1-\cos^2(\theta_0)}{1-\cos^2(\theta(t))}$

Hence 

$ \dot \varphi(t)=\dot{\varphi}_0\frac{1-{u_0}^2}{1-u(t)^2}$ where $u(0)=u_0=\cos(\theta_0)$

Analytically the solutions are 

$u(t)=A\sin(\Omega_0 t)+B \cos(\Omega_0 t)$

at;     $t= 0 \hspace{1cm} u(0)=u_0$  So $B=u_0$

and we also know that $\dot u(t)=-\sin\theta(t)\frac{d\theta(t)}{dt}$

so $\dot u(0)=-\sin(\theta_0)\dot\theta_0={-\sqrt{{1-u_0}^2}}\dot\theta_0$

Now $\dot u(t)=A\Omega_0\cos(\Omega_0t)-B\Omega_0\sin(\Omega_0t)$

at;   $t=0 \hspace{1cm} \dot u(0)={-\sqrt{{1-u_0}^2}}\dot\theta_0$ So $A\Omega_0={-\sqrt{{1-u_0}^2}}\dot\theta_0$

Hence $A = \frac{{-\sqrt{{1-u_0}^2}}\dot\theta_0}{\Omega_0}=\frac{{-\sqrt{{1-u_0}^2}}\dot\theta_0}{\sqrt{\dot{\theta}_0^2 +\dot{\varphi}_0^2 \sin^2(\theta_0)}}$

So the general solution is

$u(t)=\frac{-\sqrt{{1-u_0}^2}\dot\theta_0}{\sqrt{\dot{\theta}_0^2 +\dot{\varphi}_0^2 \sin^2(\theta_0)}}\sin(\Omega_0 t)+u_0 \cos(\Omega_0 t)$
