This worksheet will illustrate the main features of the forced oscillator. At sufficiently large times, once the effect of initial conditions are gone, an external source in the form

$A_0 \cos(\omega t)$

will result in an oscillation in the form

$
x(t) = \frac{A_0}{m((\omega_i^2-\omega^2)^2+\tilde{\gamma}^2\omega^2)}\left[(\omega_i^2-\omega^2)\cos(\omega t)+\tilde{\gamma} \omega \sin(\omega t)\right]
$

Here, $\omega_i^2 = \frac{k}{m}$, $\tilde{\gamma} = \frac{\gamma}{m}$.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def field(t, omega, a_0=1):
    """ source
    """
    return a_0*np.cos(omega*t)

def x_t(t, omega, omeg_i, gam):
    """response function, defined so that A_0/m = 1
    """
    
    w_coef = (omeg_i**2-omega**2)
    g_coef = gam*omega
    a_coef = 1.0/(w_coef**2+g_coef**2)
    return a_coef*(w_coef*np.cos(t*omega)+g_coef*np.sin(t*omega))
```

We set up parameters such that

$\omega_i = 4$
$\tilde{\gamma} = 0.5$


```python
omega_i = 4.0
gamma_tilde = 0.5
```

First case, $\omega < \omega_i$

```python
omega = 2.0
t_end = 24
num_pnts = 200
t = np.linspace(0, t_end, num_pnts)
```

```python
plt.plot(t, field(t, omega, 0.05), 'r')
plt.plot(t, x_t(t, omega, omega_i, gamma_tilde), 'g')
plt.show()
```

Clearly, the spring oscillates almost in phase with the source.

Next case $\omega - \omega_i$ positive but small

```python
omega = 3.90
t_end = 24
num_pnts = 200
t = np.linspace(0, t_end, num_pnts)
```

```python
plt.plot(t, field(t, omega, 0.05), 'r')
plt.plot(t, x_t(t, omega, omega_i, gamma_tilde), 'g')
plt.show()
```

There is a phase shift, and the amplitude of the response increases considerably.

Next case $\omega = \omega_i$, at the resonance.

```python
omega = 3.90
t_end = 24
num_pnts = 200
t = np.linspace(0, t_end, num_pnts)
plt.plot(t, field(t, omega, 0.05), 'r')
plt.plot(t, x_t(t, omega, omega_i, gamma_tilde), 'g')
plt.show()
```

Past the resonance $\omega > \omega_i$, it is away from the resonance.

```python
omega = 5.0
t_end = 24
num_pnts = 200
t = np.linspace(0, t_end, num_pnts)
plt.plot(t, field(t, omega, 0.05), 'r')
plt.plot(t, x_t(t, omega, omega_i, gamma_tilde), 'g')
plt.show()
```

The amplitude reduces considerably, but note that the reponse is at a phase shift of $\pi$ with the source

Let us plot the phase shift

```python
def phase_shift(omega, omeg_i, gam):
    return np.arctan(gam*omega/(omeg_i**2-omega**2))
def heat(omega, omeg_i, gam):
    return np.pi*np.abs(np.sin(phase_shift(omega, omeg_i, gam)))/np.sqrt((omega**2-omeg_i**2)**2+gam**2*omega**2)
```

```python
om = np.linspace(0.0, 5*omega_i, 100)
plt.plot(om, phase_shift(om, omega_i, gamma_tilde))
plt.show()
```

```python
om = np.linspace(1e-5, 2*omega_i, 300)
plt.plot(om, heat(om, omega_i, gamma_tilde))
plt.show()
```
