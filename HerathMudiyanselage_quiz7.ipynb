Correspoding hamiltonian is

$H=\frac{P^2}{2m}+\frac{1}{2}m \omega^2 x^2 + mgx \sin(\omega t)$

Now the equations of motion are 

$\dot x= \frac{p}{m} =p/2$

$\dot p = - m\omega x-mgsin(\omega t)=-4x-19.6sin(2t)$

$\dot x = -2x-9.8sin(2t)$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f_int(x,t):
    return [x[0]/2]
```

```python
x_0 = np.array([1])
t = np.linspace(0.0, 10.0, 100)
```

```python
x_sol = odeint(f_int, x_0, t)
```

```python
plt.plot(t, x_sol[:, 0])
plt.show()
```

```python
def f_int(x,t):
    return [-2*x[0]-9.8*np.sin(2*t)]
```

```python
x_0 = np.array([0])
t = np.linspace(0.0, 10.0, 100)
```

```python
x_sol = odeint(f_int, x_0, t)
```

```python
plt.plot(t, x_sol[:, 0])
plt.show()
```
