```python
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def hh_func(x, y):
    return 0.5*(x**2+y**2) + (x**2*y-y**3/3.0)

num_pnts = 500

x_min = -np.sqrt(3)/2.0
x_max = np.sqrt(3)/2.0
y_min = -1/2.0
y_max = 1.0
n_of = 0.0
v_mat = np.linspace(0.01, 1/6.0, 10)

 
# x_min= -3
# x_max=  3
# y_min= -3
# y_max= 3
# n_of = -5
```

```python
%matplotlib notebook
fig = plt.figure()
ax = fig.gca(projection='3d')

x = np.linspace(x_min, x_max, num_pnts)
y = np.linspace(y_min, y_max, num_pnts)

X , Y = np.meshgrid(x, y)

Z = hh_func(X, Y)

# Plot the 3D surface
ax.plot_surface(X, Y, Z, alpha=0.9)

# Plot projections of the contours for each dimension.  By choosing offsets
# that match the appropriate axes limits, the projected contours will sit on
# the 'walls' of the graph
cset = ax.contour(X, Y, Z, v_mat, zdir='z', stride =10, offset=n_of, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=x_min, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=y_min, cmap=cm.coolwarm)

#ax.set_xlim(-40, 40)
#ax.set_ylim(-40, 40)
#ax.set_zlim(-100, 100)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Energy')

plt.show()
```
