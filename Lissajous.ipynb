Here we will discuss the properties of Lissajous figures, defined by

$
x(t) = A\sin(a t + \delta)
$

$
y(t) = B\sin(b t)
$


```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def liss(t, a, b, amp_a, amp_b, delt):
    """returns the Lissajous figure for the given paramters
    """
    return np.array([amp_a*np.sin(a*t+ delt), amp_b*np.sin(b*t)])
```

First case, Isotropic oscillator: $A=B=1, a=b=1$, several phases $\delta$

```python
num_pnts = 200
delt_num = np.array([0.0, np.pi/4, np.pi/2, 3*np.pi/4.0])
num_delt = delt_num.shape[0]
a= 1.0
amp_a = 1.0
b = a
amp_b = amp_a
temp = np.linspace(0.0, 2*np.pi*5, num_pnts)
```

```python
for ind in range(num_delt):
    plt.subplot(2, 2, ind+1)
    x, y = liss(temp, a, b, amp_a, amp_b, delt_num[ind])
    plt.plot(x, y)
    plt.text(-1.5, 0.7, r'$\delta =$')
    plt.text(-1, 0.7, '%1.2f' % delt_num[ind])
    plt.axis('equal')
    plt.xlim([-1, 1])

plt.show()
```

Second case, oscillators with fractional frequencies $a/b=1/2$ and different phases

```python
num_pnts = 200
delt_num = np.array([0.0, np.pi/10, np.pi/3, np.pi/2.0])
num_delt = delt_num.shape[0]
a= 1.0
amp_a = 1.0
b = 2.0*a
amp_b = amp_a
temp = np.linspace(0.0, 2*np.pi*5, num_pnts)
for ind in range(num_delt):
    plt.subplot(2, 2, ind+1)
    x, y = liss(temp, a, b, amp_a, amp_b, delt_num[ind])
    plt.plot(x, y)
    plt.text(-1.5, 0.7, r'$\delta =$')
    plt.text(-1, 0.7, '%1.2f' % delt_num[ind])
    plt.axis('equal')
    plt.xlim([-1, 1])

plt.show()
```
