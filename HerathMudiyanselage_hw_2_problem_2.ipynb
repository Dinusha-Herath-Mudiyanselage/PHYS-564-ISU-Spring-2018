First import all the packages necessary for the program.

```python
import numpy as np                   #import the numpy package
from scipy.integrate import odeint   #import the odient function from scipy.integrate module
import matplotlib.pyplot as plt      #import the matplotlib package
```

Next define the diffential equation.  

$ \frac {d^2y}{dx^2}+y=0$ 

with $y(0)=0$ and $y \prime (0)=1$

We can write the above differential eqaution with two other eqautions.

$\frac{dy_1}{dx}=-y_2$

$ \frac{dy_2}{dx}=y_1$  

with $y_1(0)=0$ and $y_2(0)=-1$


```python
def f_int(y,t):            #define the ODE
    return [-y[1],y[0]]
```

Then implement initial conditions

```python
y_0 = np.array([0.0,-1.0])       # an array with initial conditions 0 and -1
t = np.linspace(0.0, 20.0, 100)  # create a line space array with 100 eqaul space points between 0 and 20
```

After that solve the differential equation

```python
y_sol = odeint(f_int, y_0, t)    #solve the diffential equation with boundary conditions
```

For the differential equation $ \frac {d^2y}{dx^2}+y=0$ 

Analitically the general solution is $y(x)=Asin(x)+Bcos(x)$

Applying initial conditions

$y(0)=0=B$ and $y \prime (0)=1=A$

Hence the exact solution is $y(x)=sin(x)$

Next define the exact solution

```python
def y_sol_exact(x):            #define the sin function
    return np.sin(x)

y_sol_exct = y_sol_exact(t)    #create an array with solution values
```

Next plot both graphs in the same plot

```python
plt.plot(t, y_sol[:, 0])
plt.plot(t, y_sol_exct)
plt.show()
```

Both graphs are identical cannot distinguish between them

Next calculate the error between exact and numerical solutions

```python
plt.plot(t, 100*np.abs((y_sol[:, 0]-y_sol_exct)/y_sol_exct))
plt.show()
```
