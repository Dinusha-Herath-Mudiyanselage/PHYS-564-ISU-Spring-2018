```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
beta=3/4
omega=3/2

def f_int(q,t):
    return [-beta*q[0]-omega**2*q[1],q[0]]
```

```python
y_0 = np.array([0,-omega**2*1.7])
t = np.linspace(0.0, 20.0, 100)
```

```python
y_sol = odeint(f_int, y_0, t)
```

```python
plt.plot(t, y_sol[:, 0])
plt.show()
```
