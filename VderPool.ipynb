In this worksheet we are going to analyze the Van Der Pool equation, defined as

$
\frac{d u}{dt} = \varepsilon(1-v^2)u-v + F \cos(\omega_r t)
$

$
\frac{d v}{dt} = u
$

```python
import numpy as np
import matplotlib.pylab as plt
from scipy.integrate import odeint
```

```python
def fun(y, t, eps, fo=0, omeg=0.0):
    """
    Integration function
    """
    f1 = y[1]
    f2 = eps*(1-y[0]**2)*y[1] - y[0] + fo*np.cos(omeg*t)
    
    return np.array([f1, f2])
```

```python
def solv(y0, t, eps, fo=0, omeg=0.0):
    return odeint(fun, y0, t, args=(eps, fo, omeg))
```

Let us plot the case $\varepsilon = 0$

```python
eps = 0.0
t_end = 8.0*np.pi
num_points = 100
y0 = np.array([2.0, 0])
temp = np.linspace(0.0, t_end, num_points)
solv_eps_0 = solv(y0, temp, eps)
plt.plot(temp, solv_eps_0)
plt.show()
```

```python
plt.plot(solv_eps_0[:, 0], solv_eps_0[:, 1])
plt.show()
```

The canDerPool equation clearly has a 1d attractor. Now we will analyze if solutions show chaos. For this purpose, we will chose a solution with very close initial conditions

```python
eps = 4.0
num_iter = 10
t_end = 2.0*np.pi*num_iter
num_points = 100*num_iter
temp = np.linspace(0.0, t_end, num_points)

y0 = np.array([4.0, 0])
y1 = np.array([4.1, 0])

solv_0 = solv(y0, temp, eps)
solv_1 = solv(y1, temp, eps)
plt.plot(temp, np.abs(solv_0[:, 0]-solv_1[:, 0]))
plt.show()
```

Clearly, these solutions are NOT chaotic.
