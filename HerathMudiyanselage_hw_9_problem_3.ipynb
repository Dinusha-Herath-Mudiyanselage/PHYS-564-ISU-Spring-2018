The kinetic energy of the system

$T=\frac{1}{2}m\dot z ^2$

The potential energy of the system

$V=\frac{1}{2}k(z-x_{wall})^2$ where $x_{wall}=A\sin(\omega t)$

So the Lagragian

$L=T-V=\frac{1}{2}m\dot z ^2-\frac{1}{2}k(z-x_{wall})^2$

and Hamiltonian

$H=T+V=\frac{1}{2}m\dot z ^2+\frac{1}{2}k(z-x_{wall})^2$

and $p_z=\frac{\partial L}{\partial \dot z}=m\dot z$

So we can rewrite the hamiltonian

$H=\frac{p_z^2}{2m}+\frac{1}{2}k(z-x_{wall})^2$

Now the equations of motions are

$\dot z=\frac{\partial H}{\partial p_z}=\frac{p_z}{m}$  and

$-\dot p_z=\frac{\partial H}{\partial q}=k(z-x_{wall})$

Hence

$-m\ddot z=k(z-x_{wall})$

$\ddot z= -\frac{k}{m}(z-x_{wall})$ 

Let $\omega_0=\sqrt{\frac{k}{m}}$

Hence

$\ddot z +\omega_0^2z=\omega_0^2x_{wall}$

$\ddot z +\omega_0^2z=\omega_0^2A\sin(\omega t)$

Complementary solution for the homogenous eqaution $\ddot z+\omega_0^2z=0$ is

$z_c(t)=B\sin(\omega_0 t)+C\cos(\omega_0 t)$

 Patricular solution for the non homogenous eqaution $\ddot z +\omega_0^2z=\omega_0^2A\sin(\omega t)$
 
 Let $\dot z=Dz$ and $ \ddot z=D^2z$
 
 $(D^2+\omega_0^2)z=\omega_0^2A\sin(\omega t)$
 
 $z=\frac{\omega_0^2A\sin(\omega t)}{(D^2+\omega_0^2)}$
 
 $z=\frac{\omega_0^2A}{(D^2+\omega_0^2)}\frac{1}{2i}\big[e^{i\omega t}-e^{-i\omega t}]$
 
 Lets put $D=\pm i\omega$ and $D^2=-\omega^2$
 
 $z=\frac{\omega_0^2A}{(\omega_0^2-\omega^2)}\frac{1}{2i}\big[e^{i\omega t}-e^{-i\omega t}]$
 
 $z=\frac{\omega_0^2A\sin(\omega t)}{(\omega_0^2-\omega^2)}$

Now the genaral solution is

$z(t)=z_c(t)+z_p(t)=B\sin(\omega_0 t)+C\cos(\omega_0 t)+\frac{\omega_0^2A\sin(\omega t)}{(\omega_0^2-\omega^2)}$

Let $z=0$ at $t=0$

This gives $C=0$

Let $\dot z(0)=0$ at $t=0$

$0=B\omega_0\cos(\omega_0 t)+\frac{\omega_0^2\omega A\cos(\omega t)}{(\omega_0^2-\omega^2)}\big|_{t=0}$

$0=B\omega_0+\frac{\omega_0^2\omega A}{(\omega_0^2-\omega^2)}$

Hence $B=-\frac{\omega_0\omega A}{(\omega_0^2-\omega^2)}$

Now

$z(t)=\frac{\omega_0A}{(\omega_0^2-\omega^2)}\big[\omega_0\sin(\omega t)-\omega\sin(\omega_0 t)\big]$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def field(t, omega, a_0=1):
    """ source
    """
    return a_0*np.sin(omega*t)

def z_t(t, omega, omega_0):
    """response function,
    """
    
    w_coef = (omega_0**2-omega**2)
    
    
    a_coef = omega_0/w_coef
    
    return a_coef*(omega_0*np.sin(omega*t)-omega*np.sin(omega_0*t))
```

First case, $\omega < \omega_0$

```python
omega_0 = 4.0
omega = 2.0
t_end = 24
num_pnts = 200
t = np.linspace(0, t_end, num_pnts)
```

```python
plt.plot(t, field(t, omega), 'r')
plt.plot(t, z_t(t, omega, omega_0), 'g')
plt.show()
```

Second case, $\omega > \omega_0$

```python
omega_0 = 2.0
omega = 4.0
t_end = 24
num_pnts = 200
t = np.linspace(0, t_end, num_pnts)
```

```python
plt.plot(t, field(t, omega), 'r')
plt.plot(t, z_t(t, omega, omega_0), 'g')
plt.show()
```

When $\omega > \omega_0$ there is phase of change of $\pi$

Next case $\omega_0 - \omega$ positive but small

```python
omega_0 = 4.0
omega = 3.9
t_end = 24
num_pnts = 200
t = np.linspace(0, t_end, num_pnts)
```

```python
plt.plot(t, field(t, omega), 'r')
plt.plot(t, z_t(t, omega, omega_0), 'g')
plt.show()
```

In this problem energy is not a constant. Energy is constantly feeding in to the system

Energy of the system is increasing with time since there is no damping.(loss off energy)

```python
omega_0=2
omega=np.linspace(1,3,300)
plt.plot(omega,np.abs(omega_0/(omega_0**2-omega**2)))
plt.xlabel('$\omega$')
plt.ylabel('Amplitude')
plt.show()
```

```python
def z_t_dot(t, omega, omega_0):
    """z_dot(t),
    """
    
    w_coef = (omega_0**2-omega**2)
    
    
    a_coef = omega_0/w_coef
    
    return a_coef*(omega_0*omega*np.cos(omega*t)-omega*omega_0*np.cos(omega_0*t))
```

```python
omega_0 = 4.0
omega = 3.9
t_end = 24
m=1
k=1
num_pnts = 200
t = np.linspace(0, t_end, num_pnts)
```
