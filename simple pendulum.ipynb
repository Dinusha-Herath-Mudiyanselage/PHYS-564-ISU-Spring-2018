Lagragian $ L=\frac{1}{2}m  l^2 (\dot\theta)  ^2 - mglsin(\theta)$

Equation of motion $\ddot\theta + \frac{g}{l}sin(\theta)=0$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
l=10 #length

def f_int(theta,t):
    return [-(9.8/l)*theta[1],theta[0]]
```

```python
y_0 = np.array([np.pi/6,0])
t = np.linspace(0.0, 20.0, 100)
```

```python
y_sol = odeint(f_int, y_0, t)
```

```python
plt.plot(t, y_sol[:, 0])
plt.plot(t,np.pi/6*np.cos(9.8/l*t))
plt.show()
```
