The Lorenz Attractor

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as la
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

The usual values, are, as discussed given by

$\sigma = 10$

$B= 8/3$

$R=28$

```python
def f(r, t, sigma=10.0, beta=8.0/3.0, r_val = 28):
  x, y, z = r  
  return np.array([sigma * (y - x), x * (r_val - z) - y, x * y - beta * z ])
```

```python
%matplotlib notebook 
r0 = [1.0, 1.0, 1.0]
t = np.arange(0.0, 40.0, 0.01)

states = odeint(f, r0, t)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(states[:,0], states[:,1], states[:,2])
plt.show()
```

Let us analyze how the solutions diverge

```python
%matplotlib inline
t_time = np.linspace(0.0, 25, 100)
y_h = 0.94
r1 = [1.0+1e-6, 1.0, 1.0]
states_1 = odeint(f, r1, t)
diff = la.norm(states_1-states, axis=1)
plt.semilogy(t, diff)
plt.semilogy(t_time, 3*1e-10*np.exp(t_time*y_h))
plt.show()
```

Thus $y_h \approx 0.94$. The second exponent is $y_s=0$ and the thrid exponent mus be $y_f \approx -12.8$

Thus the dimension is

```python
print(2+y_h/12.8)
```

From the plot, one can see that the shape of the Lorenz attractor is almost two-dimensional.
