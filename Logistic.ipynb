Here we study the Logistic equation and the emergence of bifurcations.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def log_map(x, a=3):
    """
    Logistic equation
    """
    return a*x*(1-x)

def a_fp(x):
    """
    Fixed Point
    """
    return (x-1)/x
```

Let us generate a map for a < 3.

For example, a=2.9

```python
a_param = 2.9
ini_val = 0.1

n_iter = 50
x_iter = np.zeros(n_iter)

z = ini_val
for ind in range(n_iter):
    z = log_map(z, a_param)
    x_iter[ind] = z

plt.plot(range(n_iter), a_fp(a_param)*np.ones(n_iter),'--')
plt.plot(range(n_iter), x_iter)
plt.show()
```

Let us try now a=3.2

```python
a_param = 3.2
ini_val = 0.1

n_iter = 50
x_iter = np.zeros(n_iter)

z = ini_val
for ind in range(n_iter):
    z = log_map(z, a_param)
    x_iter[ind] = z

plt.plot(range(n_iter), a_fp(a_param)*np.ones(n_iter),'--')
plt.plot(range(n_iter), x_iter)
plt.show()
```

There is a clear cycle!

Let us try the value of a = 3.45

```python
a_param = 3.45
ini_val = 0.1

n_iter = 50
x_iter = np.zeros(n_iter)

z = ini_val
for ind in range(n_iter):
    z = log_map(z, a_param)
    x_iter[ind] = z

plt.plot(range(n_iter), a_fp(a_param)*np.ones(n_iter),'--')
plt.plot(range(n_iter), x_iter)
plt.show()
```

Clearly, now we have a 4-cycle. The bifurcation points are at:
$a_2 = 3.0$

$a_4 = 3.4495$

$a_8 = 3.5441$

$a_{16} = 3.5644$

$a_{\infty} = 3.5699456$

Let us investigate $a> a_{\infty}$..

```python
a_param = 3.69
ini_val = 0.1

n_iter = 2000
x_iter = np.zeros(n_iter)

z = ini_val
for ind in range(n_iter):
    z = log_map(z, a_param)
    x_iter[ind] = z

plt.plot(range(n_iter), a_fp(a_param)*np.ones(n_iter),'--')
plt.plot(range(n_iter), x_iter, 's', markersize=2)
plt.show()
```

This is chaos

Note that the plot covers all the set, but with different probability!
