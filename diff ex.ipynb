```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f_int(y,t):
    return [-y[1],y[0]]
```

```python
y_0 = np.array([0.0,-1.0])
t = np.linspace(0.0, 20.0, 100)
```

```python
y_sol = odeint(f_int, y_0, t)
```

```python
plt.plot(t, y_sol[:, 0])
plt.show()
```

```python
plt.plot(t, y_sol[:, 0])
plt.xlim([0, 100])
plt.text(0.0,0.0, '(Many other periodic functions not plotted)')
plt.annotate('local max', xy=(0.5*np.pi, 1), xytext=(2.3, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )
plt.show()
```
